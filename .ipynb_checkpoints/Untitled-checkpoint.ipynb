```python
import os
from openai import OpenAI

client = OpenAI(api_key="YOUR_API_KEY")
```

```python
def generate_questions(skills):
    prompt = f"""
    Generate 5 technical interview questions for a candidate with skills: {skills}.
    
    For each question, provide:
    - question
    - ideal_answer
    - keywords (important concepts)
    
    Return strictly in JSON format like:
    [
      {{
        "question": "...",
        "ideal_answer": "...",
        "keywords": ["...", "..."]
      }}
    ]
    """

    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": prompt}]
    )

    content = response.choices[0].message.content

    import json
    return json.loads(content)
```

```python
from flask import Flask, request, render_template
import json
import random
import pdfplumber
from openai import OpenAI

app = Flask(__name__)   # ✅ MUST be here

client = OpenAI(api_key="YOUR_API_KEY")


# -------- FUNCTIONS --------

def extract_text_from_pdf(file):
    text = ""
    with pdfplumber.open(file) as pdf:
        for page in pdf.pages:
            text += page.extract_text() or ""
    return text.lower()


def extract_skills(text):
    skills = []
    if "python" in text:
        skills.append("Python")
    if "machine learning" in text:
        skills.append("Machine Learning")
    return skills


def generate_questions(skills):
    prompt = f"""
    Generate 3 interview questions with answers for skills: {skills}.
    Return JSON format with question, ideal_answer, keywords.
    """

    response = client.chat.completions.create(
        model="gpt-4.1-mini",
        messages=[{"role": "user", "content": prompt}]
    )

    import json
    return json.loads(response.choices[0].message.content)


# -------- ROUTES --------

@app.route("/")
def home():
    return render_template("upload.html")


@app.route("/upload", methods=["POST"])
def upload():
    file = request.files["resume"]

    text = extract_text_from_pdf(file)
    skills = extract_skills(text)

    questions = generate_questions(skills)

    return render_template("questions.html", questions=questions)


# -------- RUN --------

if __name__ == "__main__":
    app.run(debug=True)
```
